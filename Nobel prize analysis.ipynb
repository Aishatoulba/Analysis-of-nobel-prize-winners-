#### 3. Checking the missing values <a class="anchor" id="section_3_3">

```r
missing.values <- nobel %>%
    gather(key = "key", value = "val") %>%
    mutate(is.missing = is.na(val)) %>%
    group_by(key, is.missing) %>%
    summarise(num.missing = n()) %>%
    filter(is.missing==T) %>%
    select(-is.missing) %>%
    arrange(desc(num.missing)) 
missing.values %>%
  ggplot() +
    geom_bar(aes(x=key, y=num.missing), stat = 'identity') +
    labs(x='variable', y="number of missing values", title='Number of missing values') +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

#### 2.Analyzing Nobel laureates by gender <a class="anchor" id="section_4_2">
we are interested in knowing how many prizes were won by male and female. 
Since we have some missing values for the sex variable before diving into the analysis , we will ignore these values by filtering. 

```r
nobel_sex <- nobel %>% 
filter(!is.na(sex))

nobel_sex %>%
    group_by(sex) %>%
summarize(count=n())
```

##### Around 94.46% of the laureates of the nobel prize are male while 5.54% are female.



```r
ggplot(nobel_sex,aes(x=sex)) + geom_bar(color='grey',fill='black') +
ggtitle('Nobel laureates per gender') + 
xlab('Gender') +  ylab('Count') + 
coord_flip() + theme_dark()
```

#### 3.Analyzing Nobel laureates by Country <a class="anchor" id="section_4_3">

```r
nobel_country <- nobel %>% 
filter(!is.na(birth_country))

nobel_country %>%
    group_by(birth_country) %>%
summarize(count=n()) %>%
arrange(desc(count)) %>%
head(10)
```

##### The most common Nobel laureates between 1901 and 2016 were born in the United States of America.

#### 6.Analyzing Nobel laureates by Prize Category and gender <a class="anchor" id="section_4_6">

```r
nobel_sex %>%
  group_by(sex) %>%
  count(category) %>%
  ggplot(aes(x = reorder(category, -n), y = n,fill=sex)) +
  geom_bar(stat = 'identity',color='black') + 
  ggtitle('Laureates per prize category and gender') + 
  xlab("Prize category") + 
  ylab("count") + theme_dark()
```

##### For all the categories , most nobel laureates are men.It appears that there is some imbalance between how many male and female prize winners there are.Besides to that we can see that women are mostly present in the Peace category. 
##### It is very interesting to see the evolution of female winners per decade. let's see :)


#### 7.Female laureates per decade <a class="anchor" id="section_4_7">

```r
prop_female_winners <- nobel %>%
mutate(female_winner = sex =='Female') %>%
mutate(decade = floor(year/10) * 10) %>%
group_by(decade,category) %>%
mutate(proportion=mean(female_winner,na.rm=TRUE))  

# Plotting the proportion of female laureates per decade
ggplot(prop_female_winners,aes(x=decade,y=proportion,color=category))+geom_line() + geom_point() +
scale_y_continuous(labels=scales::percent,limits=0:1,expand=c(0,0)) + theme_dark()
```

##### As we can see , the imbalance is pretty large with physics , Economics and chemistry having the largest imbalance. The plot confirms also that women are pretty present in the peace category with the peace prizes increasing between 2010 and 2016

#### 8.Men laureates per decade <a class="anchor" id="section_4_8">

```r
prop_male_winners <- nobel %>%
mutate(male_winner = sex =='Male') %>%
mutate(decade = floor(year/10) * 10) %>%
group_by(decade,category) %>%
mutate(proportion=mean(male_winner,na.rm=TRUE))  

# Plotting the proportion of Men laureates per decade
ggplot(prop_male_winners,aes(x=decade,y=proportion,color=category))+geom_line() + geom_point() +
scale_y_continuous(labels=scales::percent,limits=0:1,expand=c(0,0)) + theme_dark()
```

##### We can see that all the categories have a somewhat positive trend except for peace , and literature

Around 97% of the nobel prizes handed out between 1901 and 2016 are for individuals.

#### 10. Analyzing Nobel laureates age when winning the prize ? <a class="anchor" id="section_4_10">

```r
library(lubridate)
nobel_age <- nobel %>%
mutate(age=year-year(birth_date))
```

```r
nobel_age %>% 
mutate(decade=floor(year/10)*10) %>%
group_by(decade) %>% 
summarize(average_age = mean(age,na.rm=TRUE))
```

```r
# Plotting the age of Nobel Prize winners
ggplot(nobel_age,aes(year,age)) + geom_point() + geom_smooth() + theme_dark()
```

##### We can see that people use to be around 55 when they received the prize but starting from 1990 the average is closer to 65. However we can see two outliers here , two nobel laureates below 30 years old , one is 25 years old and the another below 20s. let' see them.

```r
nobel_age %>% 
filter(age <= 25 )
```

##### The two youngest nobel laureates are : William Lawrence Bragg with 25 years old in the phyiscs category and Malala yousafzai with 17 years old in the peace category.

#### 11. Laureates age by category <a class="anchor" id="section_4_11">

```r
nobel_age_category <- nobel_age %>% 
mutate(decade=floor(year/10)*10) %>%
group_by(decade,category) %>% 
summarize(average_age = mean(age,na.rm=TRUE))
```

```r
nobel_age_category
```

```r
# Same plot as above, but faceted by the category of the Nobel Prize
ggplot(nobel_age_category,aes(decade,average_age)) + geom_point() + geom_smooth(se=FALSE) +
facet_wrap(~category) + theme_dark()
```

##### we can see that winners of chemistry , medicine , phyiscs and literature have gotten older over time.The strongest trend is within the physics category , the average age used to be below 50 and now it's almost 70.
